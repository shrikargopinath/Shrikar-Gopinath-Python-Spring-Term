The mean rating shows the average movie rating in the dataset and the standard deviation shows how spread out the ratings are from the mean. The min value shows the lowest rated movie the 75% value means that 75% of the movies have ratings below that number, and the 25% means that 25% of the movies have ratings below THAT number. Most of the movies are clustered around the average rating low variation.

```python
df['Runtime (Minutes)'].hist()
```

The histogram shows that most movie runtimes are clustered around a common range of runtime around 90 to 110 minutes. This suggests that most movies have a pretty standard length, with fewer movies being very short or abnormally long.

```python
df['Rating'].var()
```

The variance and standard deviation show how spread out the movie ratings are from the avg rating. A smaller standard deviation would mean that most movies have similar ratings, while a larger standard deviation means the movie ratings vary more. This matters because it helps show whether movie quality is consistent overall or differs largely across the dataset.
